```python
x = np.arange(0, 100.01, 0.01)
```

```python
def triangleFS(x, a, b, c):
    y = np.zeros_like(x)  # Initialize the membership degree vector with zeros
    
    # Apply membership degree rules
    y[x <= a] = 0
    y[(x > a) & (x <= b)] = (x[(x > a) & (x <= b)] - a) / (b - a)
    y[(x > b) & (x <= c)] = (c - x[(x > b) & (x <= c)]) / (c - b)
    y[x >= c] = 0
    
    return y
```

```python
y = triangleFS(x, a, b, c)
```

```python
plt.plot(x, y)
plt.title('triangleFS')
plt.xlabel('x')
plt.ylabel('Y')


plt.grid(True)
plt.show()
```

```python
def intervalFS(x, a, b):
    # Inicjalizacja wektora stopni przynależności
    y = np.zeros_like(x)

    # Obliczenia stopni przynależności
    y[x < a] = 0.0
    y[(x >= a) & (x <= b)] = 1.0
    y[x > b] = 0.0

    return y
```

```python
c, v = 30, 70  
```

```python
y = intervalFS(x, c, v)
```

```python
plt.plot(x, y)
plt.title('Stopnie przynależności w przedziale [30, 70]')
plt.xlabel('x')
plt.ylabel('Stopień przynależności')
plt.show()
```

```python
def trapezoidFS(x,a,b,c,d):
    y = np.zeros_like(x)
    
    y[x <= a] = 0
    y[(a<=x)&(x<=b)] =(x[(a<=x)&(x<=b)]-a)/(b-a)
    y[(b<=x)&(x<=c)]  =1
    y[(c<=x)&(x<=d)]=(d-x[(c<=x)&(x<=d)])/(d-c)
    y[x>=d]=0
    return y  
  
```

```python
y=trapezoidFS(x,5,50,70,90)
```

```python
plt.plot(x, y)
plt.title('Stopnie przynależności ')
plt.xlabel('x')
plt.ylabel('Stopień przynależności')
plt.show()
```

```python
def singletonFS(x,a):
    y = np.zeros_like(x)
    y[x==a]=1
    y[x!=a]=0
    return y
```

```python
def gaussFS(x, a, b):
    # Obliczenia stopni przynależności według funkcji gaussa
    y = np.exp(-(x - a)**2 / (2 * b**2))
    return y
```

```python
a, b = 60, 5
y = gaussFS(x, a, b)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Stopnie przynależności według funkcji gaussa')
ax.set_xlabel('x')
ax.set_ylabel('Stopień przynależności')

plt.show()
```

```python
def bellFS(x, a, b, c):
    # Obliczenia stopni przynależności według funkcji bell
    y = 1 / (1 + np.abs((x - c) / a)**(2 * b))
    return y
```

```python
a, b, c = 10, 14, 30
y = bellFS(x, a, b, c)
```

```python
def isEmptyFS(x, y):
    # Sprawdzenie, czy zbiór rozmyty jest pusty
    return 1 if np.all(y == 0) else 0



```

```python
def isEqualFS(x1, y1, x2, y2):
    # Sprawdzenie, czy zbiory rozmyte są sobie równe
    return int(np.array_equal(y1, y2))
```

```python
def isIncludedFS(fuzzy_set_A, fuzzy_set_B):
  

    # Sprawdź, czy stopień przynależności w A jest mniejszy lub równy stopniowi przynależności w B
    included = np.all(fuzzy_set_A <= fuzzy_set_B)

    return included.astype(int)
```

```python
def isConvexFS(y):
    n = len(y)
    


    for i in range(n):
        current_orientation = (y[(i + 1) % n] - y[i]) * (y[(i + 2) % n] - y[(i + 1) % n])

        if i == 0:
            orientation = current_orientation
        elif orientation != current_orientation:
            return 0  # Jeśli istnieje co najmniej jedna zmiana orientacji, wektor nie jest wypukły.

    return 1
```

```python
def wypuklaFunkcja(x, a, b, c):
    y = np.zeros_like(x)

    # Wzór dla zbioru wypukłego
    y[(x >= a) & (x <= b)] = 1

    return y

# Testuj wypuklą funkcję
a = 20
b = 30
c = 80
y = wypuklaFunkcja(x,a, b, c)
wynik = isConvexFS(y)
print("Czy zbiór jest wypukły:", wynik)


```

```python
def supportFS(wektor_rozmyty, x):
 

    # Find the first and last index where the membership degree is greater than zero
    non_zero_indices = np.where(wektor_rozmyty > 0)[0]

    if len(non_zero_indices) == 0:
        # If there are no non-zero membership degrees, return None
        return None, None

    # Find the start and end position of the support
    support_start = x[non_zero_indices[0]]
    support_end = x[non_zero_indices[-1]]

    return support_start, support_end
```

```python
wektor_rozmyty = triangleFS(x, 30, 50, 70)
support_start, support_end = supportFS(wektor_rozmyty, x)
```

```python
print("Obszar brzegów : ({}, {})".format(support_start, support_end))
```

```python
def coreFS(wektor_rozmyty, x):
   

    # Find the indices where the membership degree is equal to 1
    core_indices = np.where(wektor_rozmyty == 1)[0]

    if len(core_indices) == 0:
        # If there are no elements with a membership degree of 1, return None
        return None, None

    # Find the start and end position of the core
    core_start = x[core_indices[0]]
    core_end = x[core_indices[-1]]

    return core_start, core_end
```

```python
fuzzy_set = triangleFS(x, 30, 50, 70)
core_start, core_end = coreFS(fuzzy_set, x)

# Wyświetlenie wyników
print("Obszar jądra: ({}, {})".format(core_start, core_end))
```

```python
def widthFS(wektor_rozmyty, x):
  

    # Find the first and last index where the membership degree is greater than zero
    non_zero_indices = np.where(wektor_rozmyty > 0)[0]

    if len(non_zero_indices) == 0:
        # If there are no non-zero membership degrees, return None
        return None

    # Calculate the width of the fuzzy set
    width = x[non_zero_indices[-1]] - x[non_zero_indices[0]]

    return width
```

```python
def heightFS(wektor_rozmyty):
   

    # Find the maximum membership degree in the fuzzy set
    max_membership = np.max(wektor_rozmyty)

    # Calculate the height of the fuzzy set
    height = max_membership

    return height
```

```python
def relCardFS(wektor_rozmyty):
 

    # Calculate the sum of membership degrees in the fuzzy set
    sum_memberships = np.sum(wektor_rozmyty)

    # Calculate the relative cardinality
    relative_cardinality = sum_memberships / len(wektor_rozmyty)

    return relative_cardinality
```
